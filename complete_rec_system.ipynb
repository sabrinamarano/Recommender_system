<h1>POPULARITY RATINGS- MOOD</h1>

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
movies = pd.read_csv('movies.csv')
ratings = pd.read_csv('ratings.csv')

movies_ratings =movies.merge(ratings)
movies_ratings
```

```python
top_3_movies =(
movies_ratings
    .groupby(['title','genres'])
    .agg( avg_rating = ('rating','mean'), count = ('userId','count'))
    .sort_values('count', ascending=False)
    .assign(avg_rating = lambda x: round(x['avg_rating'], 0))
    .head(3)

)
top_3_movies
```

```python
def top_3_movies(movies_ratings):
    top_3_movies =(
        movies_ratings
        .groupby(['title','genres'])
        .agg( avg_rating = ('rating','mean'), count = ('userId','count'))
        .sort_values('count', ascending=False)
        .assign(avg_rating = lambda x: round(x['avg_rating'], 0))
        .head(3)
    )
    return top_3_movies
   
```

```python
top_3_movies(movies_ratings)
```

<h1>Chatbot</h1>

```python
def popularity_1st_chat_bot(movies_ratings):
    print("Hi! I'm your personal recommender. Let me recommend you some movies!")
    rec= top_3_movies(movies_ratings)
    rec= top_3_movies(movies_ratings).reset_index()
    rec =rec['title']
    return rec
    


```

```python
popularity_1st_chat_bot(movies_ratings)
```

<h1>MOOD CHATBOT</h1>

```python
def mood_movies(movies_ratings):
    print("Maybe you are in a specific mood!")
    print("Select you mood: Comedy, Drama, Adventure ,Fantasy, Thriller, Action, Horror, Musical, Mystery")
    x = input()
    movies_copy=movies_ratings[movies_ratings['genres'].str.contains(x, regex=False)].sort_values(by='rating',ascending=False)
    res=movies_copy['title'].head(3)
    print(" Top 3 movies based on your mood:")
    return res
```

<h1>ITEM REC</h1>

```python
movies_pivot=pd.pivot_table(movies_ratings, values= 'rating', index= 'userId', columns= ['movieId'],fill_value=0)
movies_pivot
```

```python
#to convert a df pivot table into a df correlation matrix
def get_corr_matrix(movies_pivot):
    movies_corr=movies_pivot.corr(method='pearson')
    return movies_corr
```

```python
movies_correlation =get_corr_matrix(movies_pivot)
movies_correlation 
```

```python
#given the correlation matrix and the title , return only the related results
def rec_movies(movies_correlation,column):
    movies_3=movies_correlation[column].sort_values(column,ascending= False).head(6).iloc[1:]
    movies_3=pd.DataFrame(movies_3).reset_index()
    res=movies_3['movieId']
    return res
```

```python
rec_movies(movies_correlation,[2])
```

```python
dictionary = movies_ratings[['movieId','title','genres']]
dictionary =dictionary.drop_duplicates()
dictionary
```

```python
#movies chatbot
def movies_chat_bot_item(movies_ratings):
    print("Do you like some recommandation based on a movie you have seen?")
    x= input()
    movies_copy=(dictionary.query('title == @x'))
    column_id =movies_copy['movieId'].reset_index()
    column=column_id['movieId'].tolist()
    movies_id_rec =rec_movies(movies_correlation,column = column).to_list()
    movies_id_rec=(dictionary.query('movieId == @movies_id_rec'))
    recom=movies_id_rec['title']
    print(f"You will probably like {recom}\n")
```

```python
movies_chat_bot_item(movies_ratings)
```

<h1>USER BASED</h1>

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
user_simil = movies_pivot
```

```python
user_similitudine = pd.DataFrame(cosine_similarity(user_simil), columns=user_simil.index, index=user_simil.index)
user_similitudine
```

```python
def rec_movies_user(user_similitudine,column):
    ex =user_similitudine[column].sort_values(column,ascending= False).iloc[1:].head(5)
    ex =pd.DataFrame(ex).reset_index()
    ex =ex['userId'].to_list()
    res= movies_ratings.query('userId == @ex').sort_values(by ='rating', ascending=False).head(5)
    res=res['title']
    return res
    
```

```python
rec_movies_user(user_similitudine,[1])
```

```python
#movies chatbot
def movies_chat_bot_user(user_similitudine):
    print("Do you wanna se what the other users like you watch? Type your user ID")
    column = []
    x = int(input())
    column.append(x)
    recom = rec_movies_user(user_similitudine,column)
    print(f"You will probably like {recom}\n")
```

```python
movies_chat_bot_user(user_similitudine)
```
